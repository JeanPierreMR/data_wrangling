```python
#Cuántos superhéroes son humanos y miden más de 2 metros (200 cms)?
```

```python
query2 = """ SELECT *
FROM heroes 
WHERE Race IN ('Human', 'Mutant')
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```

```python
#Cuántos superhéroes pesan más de 100 lbs y menos de 200 lbs?
```

```python
query2 = """ SELECT Name
FROM heroes 
WHERE Weight BETWEEN 100 AND 200
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```

```python
#Cuántos superhéroes tienen los ojos de color azul o rojo?
```

```python
query2 = """ SELECT Name, `Eye color`
FROM heroes 
WHERE `Eye color` IN ('red', 'blue')
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```

```python
#Cuántos superhéroes son Human, Mutant y tienen el pelo color Verde o son Vampiros con pelo Negro
```

```python
query2 = """ SELECT Name, Race, `Hair color`
FROM heroes 
WHERE Race IN ('Human','Mutant') AND `Hair color` IN ('Green')
OR Race IN ('Vampire') AND `Hair color` IN ('Black')
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```

```python
#Cuál es el nombre del primer superhéroe si ordeno la tabla por raza en orden descendente?
```

```python
query2 = """ SELECT Name, Race
FROM heroes 
Order by Race DESC 
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```

```python
#Cuántos superhéroes son de género masculino y cuántas de género femenino?
```

```python
query2 = """ SELECT Count(*), Gender
FROM heroes 
Group by Gender
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```

```python
#Cuántas casas publicadoras tienen más de 15 superhéroes?
```

```python
query2 = """ SELECT Count(*), Publisher
FROM heroes 
Group by Publisher
HAVING COUNT(*) > 15
"""
#id, name, Gender, Eye_color, Race, Hair_color, Height, Publisher, Skin_color, Alignment, Weight
pysqldf(query2)
```
